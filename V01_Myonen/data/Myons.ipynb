```python
# Enable inline matplotlib plotting
%matplotlib inline

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy.integrate as integrate
from uncertainties import ufloat
from uncertainties.umath import exp

# set some mpf parameters
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['figure.figsize'] = (6.0,6.0)
mpl.rcParams['figure.dpi'] = 72
mpl.rcParams['savefig.bbox'] = 'tight'
```

### Define some Functions
  * `unbinned_array` will return an unbinned dataset of binned data.
  This is useful to create some nice Histograms with custom binning.

```python
def unbinned_array(binned_array, start=0, end=None, calibration_func=lambda x: x):
    """ Returned an array with 'raw' Data for creating histograms with plt.hist
    """
    raw_data = []
    if end == None:
        end = len(binned_array)
    for x in range(start,end):
        raw_data.extend([calibration_func(x)] * binned_array[x])
    return np.array(raw_data)
```

### Load in Data and define some customization variables

```python
delay1, delay2, counts = np.genfromtxt('resolution.txt', unpack=True)
data = np.genfromtxt('myons.txt', unpack=True)
data = data[8:]
unbinned_data = unbinned_array(data)
bins = 20
offset_fit = ufloat(0.023256, 0.001)
e_fit = ufloat(0.046512, 0.001)
measurement_time = 317320.0

calibrated_data = list(map(lambda x: e_fit * x + offset_fit, data))
```

Plot and fit resolution-data with gaussian model

```python
dt = delay1 - delay2
rate = counts / 30
rate_errors = np.array([(ufloat(x, np.sqrt(x)) / 30).s for x in rate])
gaussian = lambda x, a, sigma, mu: a * np.exp(- (x-mu)**2 / (2*sigma**2))
res_coeff, res_var = curve_fit(gaussian, dt, rate)
print(
    'a = {:g}±{:g}\n'
    'σ = {:g}±{:g}\n'
    'μ = {:g}±{:g}'.format(
        res_coeff[0], np.sqrt(res_var[0][0]),
        res_coeff[1], np.sqrt(res_var[1][1]),
        res_coeff[2], np.sqrt(res_var[2][2]),
    )
)
dts = np.linspace(min(dt), max(dt), 1000)
plt.xlabel('Verzögerung [ns]')
plt.ylabel('Rate [Hz]')
plt.errorbar(dt, rate, yerr=rate_errors, fmt='r+')
plt.plot(dts, gaussian(dts, res_coeff[0], res_coeff[1], res_coeff[2]))
plt.legend(['Data', 'Fit'], loc='best')
plt.savefig('../img/resolution.pdf')
plt.show()
```

Generate a simple Histogram with `bins` Bins.

```python
y, binEdges = np.histogram(unbinned_data, bins=bins)
bincenters = 0.5 * (binEdges[1:] + binEdges[:-1])
width = 0.5 * (bincenters[-1] - bincenters[0]) / bins
medStd = np.sqrt(y)
plt.errorbar(bincenters, y, fmt='r.', yerr=medStd, xerr=width)
plt.title('Unkalibrierte Messdaten')
plt.xlabel('Kanal')
plt.ylabel('Gemessene Zerfälle / %.2f' % width)
plt.xlim(0,260)
plt.show()
```

Calculate expected number of underground-events.

```python
# number of events after discriminator- and delay-calibration
N_0 = 444
N_0 = ufloat(N_0, np.sqrt(N_0))
time = ufloat(30, 1)
f = N_0 / time

# time listening for a new event
t_stop = 10e-6
# expectation value 𝜆
𝜆 = t_stop * f

# Poisson-Distributed probability for one event
p = 𝜆 * exp(-𝜆)
print('p = {}%'.format(100 * p))
print('N_U = {}'.format(measurement_time * p))

# background-events per channel
bkg = measurement_time * p / 512
print('N_U / channel = {}'.format(bkg))
```
